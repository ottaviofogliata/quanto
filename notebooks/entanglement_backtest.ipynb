# Entanglement Backtest Notebook

```python
import sys, pathlib
sys.path.append(str(pathlib.Path('..')/'src'))
from quanto.entangle import run_entanglement_backtest
from quanto.config import load_config
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
import pandas as pd
cfg = load_config(pathlib.Path('..')/'examples/config.yaml')
tickers = ['SPY', 'QQQ', 'IWM']
res = run_entanglement_backtest(cfg, tickers, days=200, source='random')
res['entanglement']
```

```python
# Plot correlation matrix interpreted as entanglement
corr = np.array(res['correlation_matrix'])
plt.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(label='Correlation')
plt.xticks(range(len(res['tickers'])), res['tickers'])
plt.yticks(range(len(res['tickers'])), res['tickers'])
plt.title('Entanglement Heatmap')
```

```python
# Plot cumulative returns of chosen ticker vs equal-weight benchmark
prices = yf.download(' '.join(tickers), period='1y', progress=False)['Adj Close'].dropna()
returns = prices.pct_change().dropna()
strategy = returns[res['chosen']]
benchmark = returns.mean(axis=1)
cum_strategy = (1 + strategy).cumprod() - 1
cum_benchmark = (1 + benchmark).cumprod() - 1
cum_strategy.plot(label=res['chosen'])
cum_benchmark.plot(label='benchmark')
plt.ylabel('Cumulative Return')
plt.legend()
```
