# Hilbert Curve Demo Notebook

```python
import sys, pathlib
sys.path.append(str(pathlib.Path('..')/'src'))
from quanto.config import load_config
from quanto.portfolio import hilbert
import numpy as np
import matplotlib.pyplot as plt
cfg = load_config(pathlib.Path('..')/'examples/config.yaml')
basis = cfg.experiment['universe']
amps = np.full(len(basis), 1.0/len(basis))
portfolio = hilbert.HilbertPortfolio(basis, amps)
portfolio.pretty()
```

```python
# Visualize probabilities
plt.bar(basis, portfolio.probs())
plt.ylabel('Probability')
plt.title('Hilbert Portfolio Probabilities')
```
